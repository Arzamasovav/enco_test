```python
import pandas as pd

DATASET_PATH = "./test_dataset.csv"

def create_dataframe():
    """
    Считывает тестовый датасет и создает из него датафрейм.

    Returns:
        Экземпляр pd.Dataframe с данными из тестового датасета.
    """
    return pd.read_csv(DATASET_PATH, sep="|")

df = create_dataframe()

def output_numeric_metrics():
    """
    Выводит из датафрейма статистические метрики числовым столбцам.
    Метрики по столбцам object_id, lat, lon намерено не вычиcляются т.к.
    являются качественными, а не количественными, вычислять по ним метрики
    не имеет смысла.
    """
    numeric_columns = ["price", "square"]
    print(f"Общее количество записей: {len(df)}")
    for column_name in numeric_columns:
        print(f"Минимальное значение для колонки {column_name}: {df[column_name].min()}")
        print(f"Максимальное значение для колонки {column_name}: {df[column_name].max()}")
        print(f"Среднее значение для колонки {column_name}: {df[column_name].mean()}")
        print(f"Стандартное отклонение значение для колонки {column_name}: {df[column_name].std()}")

def create_dataframe():
    """Выводит тип данных для каждой колонки."""
    for col in df.columns:
        print(f"Тип колонки {col}: {df[col].dtypes}")

output_numeric_metrics()
create_dataframe()
```
